```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
np.random.seed(42)
```

```python
n_samples = 1000
```

```python
extroverts = pd.DataFrame({
    'Time_spent_Alone': np.random.normal(loc=2, scale=1, size=n_samples//2),
    'Stage_fear': np.random.normal(loc=3, scale=1, size=n_samples//2),
    'Social_event_attendance': np.random.normal(loc=8, scale=2, size=n_samples//2),
    'Going_outside': np.random.normal(loc=7, scale=1.5, size=n_samples//2),
    'Drained_after_socializing': np.random.normal(loc=3, scale=1, size=n_samples//2),
    'Friends_circle_size': np.random.normal(loc=15, scale=5, size=n_samples//2),
    'Post_frequency': np.random.normal(loc=10, scale=3, size=n_samples//2),
    'Personality': 1
})
```

```python
introverts = pd.DataFrame({
    'Time_spent_Alone': np.random.normal(loc=8, scale=2, size=n_samples//2),
    'Stage_fear': np.random.normal(loc=7, scale=1, size=n_samples//2),
    'Social_event_attendance': np.random.normal(loc=3, scale=1, size=n_samples//2),
    'Going_outside': np.random.normal(loc=3, scale=1.5, size=n_samples//2),
    'Drained_after_socializing': np.random.normal(loc=7, scale=1, size=n_samples//2),
    'Friends_circle_size': np.random.normal(loc=5, scale=2, size=n_samples//2),
    'Post_frequency': np.random.normal(loc=3, scale=2, size=n_samples//2),
    'Personality': 0
})
```

```python
df = pd.concat([extroverts, introverts], ignore_index=True)
```

```python
for col in df.columns[:-1]:
    df[col] = df[col].clip(lower=0)
```

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
X = df.drop('Personality', axis=1)
y = df['Personality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Classification Accuracy:", accuracy)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.decomposition import PCA
```

```python
np.random.seed(42)
n_samples = 1000
```

```python
extroverts = pd.DataFrame({
    'Time_spent_Alone': np.random.normal(2, 1, n_samples//2),
    'Stage_fear': np.random.normal(3, 1, n_samples//2),
    'Social_event_attendance': np.random.normal(8, 2, n_samples//2),
    'Going_outside': np.random.normal(7, 1.5, n_samples//2),
    'Drained_after_socializing': np.random.normal(3, 1, n_samples//2),
    'Friends_circle_size': np.random.normal(15, 5, n_samples//2),
    'Post_frequency': np.random.normal(10, 3, n_samples//2),
    'Personality': 1
})
```

```python
introverts = pd.DataFrame({
    'Time_spent_Alone': np.random.normal(8, 2, n_samples//2),
    'Stage_fear': np.random.normal(7, 1, n_samples//2),
    'Social_event_attendance': np.random.normal(3, 1, n_samples//2),
    'Going_outside': np.random.normal(3, 1.5, n_samples//2),
    'Drained_after_socializing': np.random.normal(7, 1, n_samples//2),
    'Friends_circle_size': np.random.normal(5, 2, n_samples//2),
    'Post_frequency': np.random.normal(3, 2, n_samples//2),
    'Personality': 0
})
```

```python
df = pd.concat([extroverts, introverts], ignore_index=True)
for col in df.columns[:-1]:
    df[col] = df[col].clip(lower=0)
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
X = df.drop('Personality', axis=1)
y = df['Personality']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print("Classification Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
for column in X.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=column, hue='Personality', bins=30, kde=True, palette=['blue', 'orange'])
    plt.title(f'Distribution of {column} by Personality')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.legend(labels=['Introvert (0)', 'Extrovert (1)'])
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['Personality'] = y.values
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Personality', palette=['blue', 'orange'])
plt.title("PCA Projection of Personality Data")
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(labels=['Introvert (0)', 'Extrovert (1)'])
plt.tight_layout()
plt.show()
```
